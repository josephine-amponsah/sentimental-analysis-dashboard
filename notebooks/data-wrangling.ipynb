```python
not_needed = ['periodStartDate', 'currencyCode', 'reportGranularity', 'tagCodes', 'salesChannel', 'appliedAttributionModel']
```

```python
marketingData = ad_data.drop(not_needed, axis = 1)
```

```python
marketingData.head(5)
```

```python
marketingData['marketingInvestment'] = marketingData['marketingInvestment'].str.replace(',', '')
marketingData['marketingInvestment'] = marketingData['marketingInvestment'].astype(int)
```

```python
marketingData = marketingData.groupby(['periodCode', 'platformCode', 'channelCode']).sum(['marketingInvestment', 'impressions', 'clicks', 'visits', 'conversions', 'deliveries'])
```

```python
marketingData.head()
```

```python
marketingData = marketingData.reset_index()
```

```python
unnecessary = ['tagCodes', 'adCode', 'adGroup', 'adName', 'adContent', 'adSearchQuery', 'adProductCode', 'profit', 'currencyCode']
salesData = salesData.drop(unnecessary, axis=1)
```

```python
salesData.head()
```

```python
revenueData = salesData[['createdAt', 'platformCode', 'channelCode','revenue']]
```

```python
revenueData['createdAt'] = revenueData['createdAt'].str.replace('T00:00:00', '')
```

```python
revenueData = revenueData.groupby(['createdAt', 'platformCode','channelCode']).sum('revenue')
```

```python
revenueData = revenueData.reset_index()
revenueData.head()
```
